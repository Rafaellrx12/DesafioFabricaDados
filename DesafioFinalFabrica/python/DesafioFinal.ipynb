Existem dados nulos no banco? Se sim, quais colunas apresentam e quantos são os dados nulos nelas?



```python
import pandas as pd;
df = pd.read_csv("dados_rh.csv");


if df.isnull().values.any():

    dados_nulos = df.isnull().sum()

    colunas_com_nulos = dados_nulos[dados_nulos > 0]


    display("Colunas com dados nulos e quantidade de nulos nelas:")
    display(colunas_com_nulos)
else:
    display("Não tem dados nulos no Banco");



```

Questão 2

Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("dados_rh.csv")
df2 = df.groupby('Department')[['EmployeeNumber']].count()
df2.plot(kind='bar')
plt.title("Número de Funcionários por Departamento");

plt.show()
```

Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);


```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("dados_rh.csv")
df2 = df['Education'].value_counts().sort_index();


df2.plot(kind='line', marker='o');


plt.title('Quantidade de Funcionários por Escolaridade')
plt.xlabel('Escolaridade')
plt.ylabel('Quantidade de Funcionários')


plt.show()
```

Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("dados_rh.csv")

df2 = df['Attrition'].value_counts();
df2.plot(kind='pie', autopct='%1.1f%%')


plt.title('Relatório de funcionários com atrito');
plt.xlabel('Atrito')
plt.ylabel('');

plt.show()
```
